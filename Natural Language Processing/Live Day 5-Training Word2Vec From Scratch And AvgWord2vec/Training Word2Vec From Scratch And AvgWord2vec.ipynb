```python
# importing datasets
import pandas as pd

message = pd.read_csv('SMSSpamCollection', sep='\t',names=["label", "message"])
```

```python
# data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]',' ',message['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
# creating the bag of words model
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500,binary=True,ngram_range=(2,2))
x = cv.fit_transform(corpus).toarray()
```

```python
x[1]
```

```python
x.shape
```

```python
y = pd.get_dummies(message['label'])
y = y.iloc[:,1].values
```

```python
# train test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB

spam_detect_model = MultinomialNB().fit(x_train,y_train)
y_pred = spam_detect_model.predict(x_test)
```

```python
# prediction
y_pred = spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

score = accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

## Word2Vec implementation

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['message'][i])  # Replacing non-alphabetic characters with space
    review = review.lower()  # Converting to lowercase
    review = review.split()  # Splitting into words
    # Applying stemming and filtering out stopwords
    review = [stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]
    corpus.append(' '.join(review))  # Joining back to form the sentence
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess 
```

```python
corpus[0]
```

```python
words=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for word in simple_preprocess(sent):
        for sent in sent_token:
            words.append(simple_preprocess(sent))
```

```python
import gensim
```

```python
# lets tarin word2vec mode

w2v_model = gensim.models.Word2Vec(words,window=5,min_count=2)
```
